```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='constant',fill_value=0)
imp.fit([[1,2],[np.nan,3],[7,np.nan]]) #calls fit() on training data

x = [[np.nan,2],[6,np.nan],[7,np.nan],[6,np.nan]]# test data
print(imp.transform(x))
```

```python
df = pd.DataFrame([['a','x'],
                   [np.nan,'y'],
                   ['a',np.nan],
                   ['b','y']], dtype='category')


imp  = SimpleImputer(strategy='most_frequent')
print(imp.fit_transform(df))
print(type(df)  )
```
