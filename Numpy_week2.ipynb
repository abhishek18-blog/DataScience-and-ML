```python
import numpy as np

# Create a base array
arr = np.array([1, 2, 3, 4, 5])

# Create a view of the array
view_arr = arr[2:4]

# Create a copy of the array
copy_arr = arr.copy()

# Check if they are views or copies
print("arr is a view or copy?", arr.base is None)  # arr is original, so it has no base
print("view_arr is a view?", view_arr.base is arr)  # Should print True, as view_arr shares data with arr
print("copy_arr is a copy?", copy_arr.base is arr)  # Should print False, as copy_arr has its own data
```
