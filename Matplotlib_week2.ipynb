```python
from matplotlib import style
style.use("ggplot")

x = np.array([14,33,33,42,53,60,76])
y = np.array([40,45,50,55,65,75,85])

x1 = np.array([11,24,33,49,52,62,73])
y1 = np.array([40,45,50,55,65,75,85])

plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Study Hours vs Test Scores")

plt.plot(x,y,label="student 1")
plt.plot(x1,y1,label="student 2")

plt.legend()
plt.show()

```

```python
study_hours = np.array([1, 2, 3, 4, 5])
test_scores = np.array([40, 55, 65, 75, 85])
sleep_hours = np.array([8, 7.5, 7, 6.5, 6])

plt.subplot(1, 2, 1) # Plot 1: Study Hours vs Test Score
plt.plot(study_hours, test_scores, marker='o', color='blue')
plt.title("Study vs Score")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")

plt.subplot(1, 2, 2) # Plot 2: Sleep Hours vs Test Score
plt.plot(sleep_hours, test_scores, marker='s', color='green')
plt.title("Sleep vs Score")
plt.xlabel("Sleep Hours")
plt.ylabel("Test Score")
plt.show()
```
